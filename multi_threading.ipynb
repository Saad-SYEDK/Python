<h3>Multithreading</h3>

Multithreading is a technique that allows multiple threads to run concurrently within a single process


Lets create two funtions that takes some time to execute, we will execute them without multithreading 
and see how long it takes

we are using time.sleep() to replicate any waiting time in real life, maybe waiting for any api response from a server or waiting of i/o response from the user, etc

```python
import time   

def calculate_square(numbers):
    for n in numbers:
        time.sleep(0.2)
        print('Square:', n * n)

def calculate_cube(numbers):
    for n in numbers:
        time.sleep(0.2)
        print('Cube:', n * n * n)


numbers = [1, 2, 3, 4, 5]

start = time.time()
calculate_square(numbers)
calculate_cube(numbers)
end = time.time()

print("Total time: ", end-start)
```

Now let us use multithreading to do the same task   


```python
import time   
from threading import Thread

def calculate_square(numbers):
    for n in numbers:
        time.sleep(0.2)
        print('Square:', n * n)

def calculate_cube(numbers):
    for n in numbers:
        time.sleep(0.2)
        print('Cube:', n * n * n)


numbers = [1, 2, 3, 4, 5]

start = time.time()

# Creating two threads
t1 = Thread(target=calculate_square, args=(numbers, ))
t2 = Thread(target=calculate_cube, args=(numbers, ))

t1.start()
t2.start()

# Wait till the thread is executed completely
t1.join()
t2.join()

end = time.time()

print("Total time: ", end-start)
```

As we can see, the difference is roughly double.

One more thing you can notice in the output is how the different statments are intermixing, this is because the print statments of different threads are executing at the same time casuing the mixed output. 

This is how powerful multithreading is.

```python
import threading

def sleepMe(i):
    print("Thread %i will sleep." % i)
    time.sleep(5)
    print("Thread %i is awake" % i)

for i in range(10):
    th = Thread(target=sleepMe, args=(i, ))
    th.start()
    print("Current Threads: %i." % threading.active_count())
```
